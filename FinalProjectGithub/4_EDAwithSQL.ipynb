## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
# get base table info 
%sql PRAGMA table_info('SPACEXTABLE');
```

```python
# get just the column names
table_info = %sql PRAGMA table_info('SPACEXTABLE');
columns = [row[1] for row in table_info]
columns
```

```python
# now let's get the different values for 'Launch_Site'
sites = %sql select distinct "Launch_Site" from SPACEXTABLE;
sites
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
launch_cca = %sql select * from SPACEXTABLE where Launch_Site like '%CCA%' limit 5;
launch_cca
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
# lets get the sum of all PAYLOAD_MASS__KG_ where Launch_Site like 'KSC' (CRS ISN'T A THING?!?!?)
total_payload = %sql select sum(PAYLOAD_MASS__KG_) from SPACEXTABLE where Launch_Site like '%KSC%';
total_payload
```

```python
%sql select Payload, PAYLOAD_MASS__KG_, Launch_Site from SPACEXTABLE where Launch_Site like '%KSC%';
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
# get distinct values of booster version
boosters = %sql select distinct Booster_Version from SPACEXTABLE;
boosters
```

```python
# ok now get all F9 v1.1
boosters_f9v11 = %sql select * from SPACEXTABLE where Booster_Version like '%F9 v1.1%';
boosters_f9v11
```

```python
# now get average value of booster_version for these rows
ave_payload_f9v11 = %sql select avg(PAYLOAD_MASS__KG_) from SPACEXTABLE where Booster_Version like '%F9 v1.1%';
ave_payload_f9v11
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
# first figure out what type of outcomes there are for Landing_Outcome
outcomes = %sql select distinct Landing_Outcome from SPACEXTABLE;
outcomes
```

```python
# get min(date) where Landing_Outcome = 'Success (ground pad)'
min_date = %sql select min(Date) from SPACEXTABLE where Landing_Outcome = 'Success (ground pad)';
min_date
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
# Landing_Outcome = 'Success (drone ship)'
# where PAYLOAD_MASS__KG_ BETWEEN 4000 AND 6000
# distinct Booster_Version
booster_success = %sql select distinct Booster_Version from SPACEXTABLE where PAYLOAD_MASS__KG_ between 4000 and 6000;
booster_success
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
# ok so we want to get a count of successful and failure outcomes
# let's first look at what the distinct outcomes are
mission_outcomes = %sql select distinct Mission_Outcome from SPACEXTABLE;
mission_outcomes
```

```python
# ok so group by Mission_Outcome and see what it gives
mission_total = %sql select Mission_Outcome, Count(*) as Count from SPACEXTABLE group by Mission_Outcome order by count desc;
mission_total
```

### Task 8



##### List all the booster_versions that have carried the maximum payload mass, using a subquery with a suitable aggregate function.


```python
# I want to make a subquery that selects the max payload mass
# then select all booster versions that have that value
booster_maxpayload = %sql select Booster_Version from SPACEXTABLE where PAYLOAD_MASS__KG_ = (select max(PAYLOAD_MASS__KG_) from SPACEXTABLE);
booster_maxpayload
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
# get months using substr(Date, 6, 2) as month
# subst(Date, 0, 5) = '2015' for year
records_months_2015 = %sql select substr(Date, 6, 2) as month, Booster_Version, Landing_Outcome, Launch_Site, substr(Date,0,5) as year from SPACEXTABLE where substr(Date, 0,5) = '2015' and Landing_Outcome = 'Failure (drone ship)';
records_months_2015
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
# group by and count but now where dates are between two times
outcomes_total_dates = %sql select Landing_Outcome, Count(*) as Count from SPACEXTABLE where Date between '2010-06-04' and '2017-03-20' group by Landing_Outcome order by count desc;
outcomes_total_dates
```
